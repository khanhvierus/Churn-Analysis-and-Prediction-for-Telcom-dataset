### We found some negative values in column 'monthly_charge' . Let's check how many such values there are in the whole dataset 

```python
%%sql 
select 
 count(*) filter (where monthly_charge < 0) as monthly_charge, 
 count(*) filter (where total_charges < 0) as total_charges, 
 count(*) filter (where total_extra_data_charges < 0) as total_extra_data_charges, 
 count(*) filter (where total_long_distance_charges < 0) as total_long_distance_charges , 
 count(*) filter (where total_refunds < 0) as total_refund , 
 count(*) filter (where total_revenue < 0) as total_revenue
from customer_churn ;
```

### There are 107 negative values in column 'monthly_charge' only , as this is a small amount , we decide to delete them all  

```python
%%sql 
delete from customer_churn 
where monthly_charge < 0 ; 
```

### Add a new column 'Churn' from 'customer_status' 

```python
%%sql 
alter table customer_churn 
add column churn float ; 

update customer_churn 
set 
 churn = case 
   when customer_status = 'Churned' then 1 
   else 0 
  end; 
```

### Add new column 'tenure_category' to distinguish between new customers , long-term customers , ... 

```python
%%sql 
alter table customer_churn 
add column tenure_category text ; 

update customer_churn 
 set tenure_category = case 
   when tenure_in_months <= 12 then 'New customer'
   when tenure_in_months between 13 and 24 then 'Mid-term customer'
   when tenure_in_months > 24 then 'Long-term customer'
  end; 
```

# 3. QUERIES 

### 1. Sum of churns 


```python
%%sql 
select sum(churn)
from customer_churn ; 
```

### 2. Churn rate by gender 

```python
%%sql 
select gender , count(customer_id) as total , round( sum(churn)::numeric/count(customer_id) , 3) as churn_rate , sum(churn)
from customer_churn 
group by gender;
```

### 3. Churn Rate by Marital Status

```python
%%sql 
select married , count(customer_id) as total , round( sum(churn)::numeric/count(customer_id) , 2) as churn_rate
from customer_churn
group by married ; 
```

### 4. Churn rate by state 

```python
%%sql 
select state , count(customer_id) as total , round( sum(churn)::numeric/count(customer_id) , 2) as churn_rate
from customer_churn
group by state
order by churn_rate desc ; 
```

### 5. Top 5 states which got the most referrals to our company 

```python
%%sql 
select state , count(number_of_referrals) as total_referrals 
from customer_churn 
group by state 
order by total_referrals desc 
limit 5 ;
```

### 6. Churn rate by customer loyalty 

```python
%%sql 
select tenure_category , count(customer_id) as total , round( sum(churn)::numeric/count(customer_id) , 2) as churn_rate
from customer_churn
group by tenure_category
order by churn_rate desc ;
```

### 7. Churn rate by value deal 

```python
%%sql 
select value_deal , count(customer_id) as total , round( sum(churn)::numeric/count(customer_id) , 2) as churn_rate
from customer_churn
group by value_deal
order by churn_rate desc ;
```

### 8. Churn rate by service groups : 

```python
%%sql 
select 'phone_service' as service, phone_service as status,
       count(customer_id) as total,
       round(sum(churn)::numeric / count(customer_id), 2) as churn_rate
from customer_churn
group by phone_service

union all
select 'multiple_lines', multiple_lines,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by multiple_lines

union all
select 'internet_service', internet_service,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by internet_service

union all
select 'online_security', online_security,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by online_security

union all
select 'online_backup', online_backup,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by online_backup

union all
select 'device_protection_plan', device_protection_plan,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by device_protection_plan

union all
select 'premium_support', premium_support,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by premium_support

union all
select 'streaming_tv', streaming_tv,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by streaming_tv

union all
select 'streaming_movies', streaming_movies,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by streaming_movies

union all
select 'streaming_music', streaming_music,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by streaming_music

union all
select 'unlimited_data', unlimited_data,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by unlimited_data

union all
select 'paperless_billing', paperless_billing,
       count(customer_id),
       round(sum(churn)::numeric / count(customer_id), 2)
from customer_churn
group by paperless_billing;
```

# 4. Summary of Findings

- Churn rate does not differ between gender and marital status.  

- The state with the highest Churn rate is **Jammu & Kashmir** (0.57), while **Uttar Pradesh** has the most Referrals (4682).  

- For the group of reasons for leaving the service:  
  - The majority of customers did not leave a specific reason (4778).  
  - The most common explicit reason is **Competitor** (752).  

- In the following service groups, the Churn rate of the **registered group** is higher than the non-registered group:
  - Internet Service  
  - Multiple Lines (multiple phone numbers at the same time)  
  - Paperless Billing (electronic bill)  
  - Phone Service  
  - Streaming Movies  
  - Streaming Music  
  - Streaming TV  
  - Unlimited Data  

- On the contrary, for the following service groups, the Churn rate of the **non-registered group** is higher than the registered group:
  - Device Protection Plan  
  - Online Backup  
  - Online Security  
  - Premium Support  

- For Age groups:  
  - The majority are from **20–59 years old**.  
  - The highest Churn rate is **0.36** in the elderly group (≥ 60).  

- For promotional packages:  
  - The Churn rate increases gradually with the number of Deals.  
  - The higher the Deal, the higher the Churn rate.  
  - The highest is **Deal 5** with a Churn rate of **0.54**.  

- According to the Tenure classification:  
  - The **New customers** group (≤ 12 months) has the highest number.  
  - The Churn rate across all 3 Tenure groups is quite similar.  
  - Surprisingly, the **Long-term customers** group (> 24 months) has the highest Churn rate (**0.28**).  

- Regarding the Internet network classification:  
  - The most used type is **Fiber Optic** (2725).  
  - It also has the highest Churn rate (**0.41**).  

- Regarding the Contract type:  
  - **Month-to-month** contracts account for the majority and also have the highest Churn rate (**0.46**).  
  - The Churn rate decreases as the contract term increases, which is reasonable.  

- Regarding the Payment Method:  
  - The most popular is **Bank Withdrawal** (3521).  
  - The highest Churn rate is **Mailed Check** (0.38) with only 343 customers.  
  - **Credit Card** has many customers (2447) but the lowest Churn rate (**0.15**).
